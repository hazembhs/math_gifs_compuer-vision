```python
import numpy as np 
import cv2 
import imageio
import matplotlib.pyplot as plt 
%matplotlib inline

def plot():
    plt.figure(figsize=(5, 5))
    plt.imshow(img, cmap='Blues')
    plt.grid()
```

```python
#for the toroid 

frames = 150
offset = 12
thetas = np.linspace(0, np.pi*2, frames)
r = 100 #radius 
t = 2 #thickness 

hex_points = np.array([
            (650,240),
            (350,240),
            (200,500),
            (350,760),
            (650,760),
            (800,500)])

with imageio.get_writer('toroid.gif', mode='I') as writer:
    for j, theta in enumerate(thetas):
        img = np.ones([1000, 1000])
        
        #draw the circles as big toroids
        for i, p in enumerate(hex_points):
            
            if j < offset: 
                dashes =  np.linspace(0, np.pi*2, 0)
            elif j < frames/2: 
                dashes =  np.linspace(0, np.pi*2, j-offset)
            else: 
                dashes =  np.linspace(0, np.pi*2, frames-j)
            for small_theta in dashes: 
                x = int(r*np.cos(small_theta) + p[0])
                y = int(r*np.sin(small_theta) + p[1])
                cv2.circle(img, (x, y), 255, 0, t-1)
         
        #draw the equilateral triangles 
        #0
        x0 = int(r*np.cos(theta) + hex_points[0][0])
        y0 = int(r*np.sin(theta) + hex_points[0][1])

        #2
        x1 = int(r*np.cos(theta) + hex_points[2][0])
        y1 = int(r*np.sin(theta) + hex_points[2][1])

        #4
        x2 = int(r*np.cos(theta) + hex_points[4][0])
        y2 = int(r*np.sin(theta) + hex_points[4][1])

        cv2.line(img, (x0, y0), (x1, y1), 0, t)
        cv2.line(img, (x2, y2), (x1, y1), 0, t)
        cv2.line(img, (x2, y2), (x0, y0), 0, t)

        #--------------------------------------------#
        theta = np.pi + theta
        #1
        x3 = int(r*np.cos(theta) + hex_points[1][0])
        y3 = int(r*np.sin(theta) + hex_points[1][1])

        #3
        x4 = int(r*np.cos(theta) + hex_points[3][0])
        y4 = int(r*np.sin(theta) + hex_points[3][1])

        #5
        x5 = int(r*np.cos(theta) + hex_points[5][0])
        y5 = int(r*np.sin(theta) + hex_points[5][1])

        cv2.line(img, (x3, y3), (x4, y4), 0, t)
        cv2.line(img, (x5, y5), (x4, y4), 0, t)
        cv2.line(img, (x5, y5), (x3, y3), 0, t)
        
        #lines between adjacent circles 
        cv2.line(img, (x0, y0), (x5, y5), 0, t)
        cv2.line(img, (x3, y3), (x0, y0), 0, t)
        cv2.line(img, (x1, y1), (x3, y3), 0, t)
        cv2.line(img, (x4, y4), (x1, y1), 0, t)
        cv2.line(img, (x2, y2), (x4, y4), 0, t)
        cv2.line(img, (x5, y5), (x2, y2), 0, t)
        
        #cross lines
        cv2.line(img, (x0, y0), (x4, y4), 0, t)
        cv2.line(img, (x2, y2), (x3, y3), 0, t)
        cv2.line(img, (x5, y5), (x1, y1), 0, t)
        

        writer.append_data(img)
        
print ('done!')
!toroid.gif
```
